### Q2-1 코딩 연습_삼각형의 면적 구하기

### 문제 : 다음은 삼각형의 밑변의 길이와 높이를 이용하여 삼각형의 면적을 구하는 프로그램이다. 올바른 실행 결과가 나오도록 밑줄 친 부분을 채우시오.

<p><실행 결과></p>
<p>삼각형의 밑변 길이 : 10</p>
<p>삼각형의 높이 : 3</p>
<p>삼각형의 면적 : 15.0</p>




````python
width = 10
height = 3

area = width * height / 2

print('삼각형의 밑변 길이 : ', width)
print('삼각형의 높이 : ', height)
print('삼각형의 면적 : ', area)
````

### Q2-2 코딩 연습_거스름돈 계산하기

### 문제 : 다음은 물건가격, 구매개수, 지불금액에 따라 거스름돈을 계산하는 프로그램이다. 밑줄 친 부분을 채우시오.

<p><실행 결과></p>
<p>물건가격 : 800원</p>
<p>구매개수 : 3개</p>
<p>지불금액 : 5000원</p>
<p>거스름돈 : 2600원</p>




````python
price = 800
buy = 3
pay = 5000

change = pay - price * buy

print('물건가격 : ', price, '원')
print('구매개수 : ', buy, '개')
print('지불금액 : ', pay, '원')
print('거스름돈 : ', change, '원')
````

### Q2-3 코딩 연습_숫자 연산자

### 다음은 프로그램의 실행결과는 어떻게 되는가? 밑줄 친 부분을 채우시오


```
x = 20
y = 4

x = y % x  # x = 4 % 20 -> x = 4(왼쪽 피연산자가 오른쪽 피연산자 보다 작으면 왼쪽 피연산자 그대로 출력
print(x)

y -= x * 2   # y = y - (x * 2) -> y = 4 - (4 * 2) -> y = -4

```





````python
x = 20
y = 4

x = y % x  # x = 4 % 20 -> x = 4(왼쪽 피연산자가 오른쪽 피연산자 보다 작으면 왼쪽 피연산자 그대로 출력
print(x)

y -= x * 2   # y = y - (x * 2) -> y = 4 - (4 * 2) -> y = -4
print(x, y)
````

### Q2-4 코딩 연습_문자열 추출과 문자열 포맷팅

### 문제 : 다음은 이름, 나이, 키를 포맷에 맞추어 출력하는 프로그램이다. 밑줄 친 부분을 채우시오.

<p><실행 결과></p>
<p>성 : 홍</p>
<p>이름 : 길동</p>
<p>홍길동의 나이 : 30세, 키 : 171.50cm</p>



````python
name = '홍길동'
age = 30
height = 171.5

print('성 : ', name[0])
print('이름 : ', name[1:])
print("%s의 나이 : %d세, 키 : %.2fcm" % (name, age, height))
````

### Q2-5 코딩 연습_키보드로 성적 입력받아 평균 구하기

### 문제 : 다음은 학생 이름과 국어, 영어, 수학 성적을 키보드로 입력받아 합계와 평균을 구하는 프로그램이다. 밑줄 친 부분을 채우시오.

<p><실행 결과></p>
<p>학생 이름을 입력하세요 : 황예린</p>
<p>국어 성적을 입력하세요 : 83</p>
<p>영어 성적을 입력하세요 : 85</p>
<p>수학 성적을 입력하세요 : 91</p>
<p>이름 : 황예린, 국어 : 83, 수학 : 91, 평균 : 86.3점</p>


````python
name = input('학생 이름을 입력하세요 : ')
kor = int(input('국어 성적을 입력하세요 : '))
eng = int(input('영어 성적을 입력하세요 : '))
math = int(input('수학 성적을 입력하세요 : '))

total = kor + eng + math
avg = total / 3

print('이름 : %s, 국어 : %d, 영어 : %d, 수학 : %d, 평균 : %.1f점' % (name, kor, eng, math, avg))
````
