### 2.1.2 변수명 규칙
1. 변수명은 영문, 숫자, 밑줄(_) 조합
2. 변수명의 기본 규칙은 영문자 대소문자, 밑줄(_), 숫자를 조합해서 사용
3. 변수명명은 숫자로 시작하면 안된다.
4. 영문자에서 대문자와 소문자는 다르게 분류된다.

### 잘못된 변수명
1. 변수명에는 @, #, $, %, ^, &, *, -, /, (, ) 등의 특수문자나 공백(' ')을 사용하면 안된다.
2. 변수명의 시작을 숫자로 시작하면 오류가 발생한다.

### 예제 2-8. 문자열의 요소 추출

````python
# 1.2.2 문자열의 데이터 형_문자열의 인덱스와 요소 추출
x = 'I am happy!'

print(x)        # x 출력
print(x[0])     # 0번 인덱스 출력
print(x[0:3])   # 0번 인덱스부터 3번 인덱스 전까지 출력
print(x[5:])    # 5번 인덱스부터 마지막 인덱스까지 출력
print(x[-1])    # 맨 마지막 인덱스 출력
print(x[-3:])   # 3번 인덱스부터 마지막 인덱스까지 출력
print(x[-4:-2]) # -4번 인덱스부터 -2번 인덱스 전까지 출력
````

### 2.4.4 문자열 포맷팅

### 예제 2-20. 문자열 포맷팅 : %s

````python
name = '김수영'
a = '나는 %s입니다.' % name  # '... %s ...' % 변수_데이터 -> 변수_데이터에 해당되는 문자열의 값이 %s 위치에 대입된다.
print(a)
````

### 예제 2-21. 문자열 포맷팅 : %d

### 예제 2-22. 정수형 숫자 0으로 채우기 : %2d

````python
year = 2020
month = 3
day = 5

a = '%d-%02d-%02d' % (year, month, day)   # %2d 전체 자릿수가 두자리이고, 자리고 남았을 때에는 0으로 채우게 된다
print(a)
````

### 예제 2-27. 문자열 포맷팅(%)를 이용한 출력

````python
score1 = 80
score2 = 87

sum = score1 + score2
avg = sum / 2

print('두 과목 점수 : %d, %d' %(score1, score2))
print('합계 : %d, 평균 : %2f' % (sum, avg))
````

### 예제 2-30. 키워드 end를 이용한 출력

````python
a = '안녕하세요'
b = '반갑습니다'

print(a)
print(b)   # print 하고 난 후 자동으로 개행

print('\n\n')   # 두 줄 개행

print(a, end = '')   # 빈 문자열로 바로 다음에 출력
print(b)
````
