### Q6-1 코딩 연습 : 튜플로 구구단표 만들기

<p>문제 : 다음은 튜플을 이용하여 구구단표를 만드는 프로그램이다. 밑줄 친 부분을 채우시오.</p>

<p><실행 결과></p>
<p>구구단표</p>
<p>------------------------------</p>
<p>2 x 1 = 2</p>
<p>2 x 2 = 4</p>
<p>2 x 3 = 6</p>
<p>...</p>
<p>2 x 9 = 18</p>
<p>------------------------------</p>
<p>3 x 1 = 3</p>
<p>3 x 2 = 6</p>
<p>3 x 3 = 9</p>
<p>...</p>
<p>9 x 8 = 72</p>
<p>9 x 9 = 81</p>
<p>------------------------------</p>

```python
dans = (2, 3, 4, 5, 6 ,7, 8, 9)

print('구구단표')
print('-' * 30)

for dan in dans:
  for i in range(1, 10):
    print('%d x %d = %d' % (dan, i, dan * i))
  print('-' * 30)
```

### Q6-3 코딩 연습 : 딕셔너리로 성적 합계/평균 구하기

<p>문제 : 다음은 딕셔너리를 이용하여 3명 학생들의 성적의 합계와 평균을 구하는 프로그램이다. 밑줄 친 부분을 채우시오.</p>

<p><실행 결과></p>
<p>김예진 : 90</p>
<p>박영진 : 95</p>
<p>김소희 : 84</p>
<p>합계 : 269, 평균 : 89.67</p>

```python
scores = {'김예진': 90, '박영진': 95, '김소희': 84}

sum = 0
for key in scores :
  sum += scores[key]

  print('%s : %d' % (key, scores[key]))

avg = sum / len(scores)

print('합계 : %d, 평균 : %.2f' % (sum, avg))
```

### Q6-5 코딩 연습 : 딕셔너리로 영어 단어 퀴즈 만들기

<p>문제 : 다음은 딕셔너리를 이용하여 영어 단어 퀴즈를 만드는 프로그램이다. 밑줄 친 부분을 채우시오.</p>

<p><실행 결과></p>
<p>사과에 해당되는 영어 단어를 입력해주세요 : apple</p>
<p>정답입니다!</p>
<p>컴퓨터에 해당되는 영어 단어를 입력해주세요 : commputer</p>
<p>틀렸습니다!</p>
<p>학교에 해당되는 영어 단어를 입력해주세요 : school</p>
<p>정답입니다!</p>
<p>책상에 해당되는 영어 단어를 입력해주세요 : tesk</p>
<p>틀렸습니다!</p>
<p>의자에 해당되는 영어 단어를 입력해주세요 : chaaar</p>
<p>틀렸습니다!</p>

```python
words = {'사과': 'apple', '컴퓨터': 'computer', '학교': 'school', '책상': 'desk', '의자': 'chair'}

for key in words:
  in_word = input('%s에 해당되는 영어 단어를 입력해주세요 : ' % key)

  if in_word == words[key]:
    print('정답입니다!')
  else:
    print('틀렸습니다!')
```
