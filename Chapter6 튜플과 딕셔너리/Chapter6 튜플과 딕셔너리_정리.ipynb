### **6.1 튜플**

### 6.1.1 튜플 생성과 추출

### 예제 6-1. 튜플 생성과 요소 추출


```
튜플명 : (데이터, 데이터, 데이터, ...)
```



````python
menu = ('coffee', 'milk', 'tea', 'cider')

print(menu)
print(menu[0])
print(menu[2])
print(menu[0:3])

menu[1] = 'cola'
````

### 6.1.2 튜플 병합과 길이

### 6-2. 튜플 병합과 길이 구하기

```
# 튜플 병합
튜플명 : 튜플1 + 튜플2 + 튜플3 + ....

# 튜플의 길이
len(튜플명)
```



````python
tuple1 = ('apple', 'banana', 'cherry')
tuple2 = ('orange', 'melon', 'strawberry')

tuple3 = tuple1 + tuple2
print(tuple3)

print(len(tuple3))

for x in tuple1:
  print(x)

del tuple1
````

### **6.2 딕셔너리**

### 6.2.1 딕셔너리의 기본 구조

### 예제 6-3. 딕셔너리의 기본 구조

```
딕셔너리명 = {키: 값, 키: 값, ...}
```



````python
members = {'name':'안지영', 'age':30, 'email':'jiyoung@korea.com'}

print(members)
print(members['name'])
print(members['age'])

print('길이 : %d' % len(members))
````

### 6.2.2 딕셔너리 요소의 추가/수정/삭제

### 예제 6-4. 딕셔너리 요소 다루기

```
# 딕셔너리 요소 추가/수정
딕셔너리명[키] = 값

# 딕셔너리 삭제
del 딕셔너리명[키]
```



````python
name  = '안진영'
scores = {'kor':95, 'eng':85, 'math':90, 'science':80}
print(scores)

scores['kor'] = 70
print(scores['kor'])

scores['mssic'] = 100
print(scores)

print('이름 : %s' % name)
print('국어 : %d' % scores['kor'])
print('영어 : %d' % scores['eng'])
print('수학 : %d' % scores['math'])
````

### 6.2.3 for문에서 딕셔너리 활용

### 예제 6-5 for문에서 딕셔너리 사용하기

```
for 변수명 in 딕셔너리명:
  ...
  딕셔너리명[변수명]
  ...
```



````python
phones = {'갤럭시 노트8': 2017, '갤럭시 S9': 2018, '갤럭시 노트10': 2019, '갤럭시 S20': 2020}
print(phones)

for key in phones:
  print('%s => %s' % (key, phones[key]))

print(len(phones))
````
