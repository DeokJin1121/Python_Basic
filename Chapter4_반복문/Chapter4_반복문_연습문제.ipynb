### 6. 1 ~ 100까지의 수 중에서 홀수와 홀수의 합을 실행 결과와 같이 출력하는 프로그램을 작성하시오.

<p>1 + 3 + 5 + ... + 99</p>

<p><실행 결과></p>
<p>1 + 3 + 5 + 7 + 9 + ... + 95 + 97 + 99 = 2500</p>

```python
sum = 0

for i in range(1, 101) :
  if i % 2 == 1 and i != 99:
    print('%d + ' % i, end = '')
    sum += int(i)
  elif i == 99 :
    print('%d ' %i, end = '')
    sum += int(i)
  elif i == 100 :
    print('= %d' % sum)
```

### 7. 다음과 같은 수식의 결과를 구하는 프로그램을 작성하시오.

<p>2^1 + 4^3 + 6^5 + ... + 2N^(2N-1)</p>

<p><실행 결과></p>
<p>N의 값을 입력하세요 : 12</p>
<p>N의 값 : 12</p>
<p>합계 : 744010475682</p>


```python
N = int(input('N의 값을 입력하세요 : '))
sum = 0

print('N의 값 : %d' % N)

for i in range(1, N + 1) :
  sum += (2 * i) ** (2 * i -1)

print('합계 : %d' % sum)
```

### 8. N 값을 입력받아 1부터 N까지의 수 중에서 소수를 구하는 프로그램을 작성하시오.

<p>* 소수 : 1과 자기 자신만으로 나누어 떨어지는 1보다 큰 양의 정수</p>

<p><실행 결과></p>
<p>2 3 5 7 11 13 ... 193 197 199</p>

```python
import math

# 소수 판별 함수(에라토스테네스의 체)
def is_prime_number(n):
  # 2부터 n까지의 모든 수에 대하여 소수 판별
  array = [True for i in range(n + 1)] # 처음엔 모든 수가 소수(True)인 것으로 초기화(0과 1은 제외)

  # 에라토스테네스의 체
  for i in range(2, int(math.sqrt(n)) + 1) : # 2부터 n의 제곱근까지의 모든 수를 확인하며
    if array[i] == True : # i가 소수인 경우(남은 수인 경우)
      # i를 제외한 i의 모든 배수를 지우기
      j = 2
      while i * j <= n :
        array[i * j] = False
        j += 1
  return [ i for i in range(2, n + 1) if array[i] ]

# N이 1,000,000 이내로 주어지는 경우 활용할 것 => 이론상 400만번 정도 연산이고 메모리도 충분함

n = int(input('n 값을 입력하세요 : '))
print(is_prime_number(n))
```

### 9.while문을 이용하여 달러($), 원화(원), 유로(€)의 환산표를 만드는 프로그램을 작성하시오.

<p>* 1$당 환율 : 1080원(원화), 0.81€(유로)</p>

<p><실행 결과></p>
<img src = "https://drive.google.com/uc?id=1VqA3FFoA89yGHrmXtKnarizYLz-ImLUp" height = 400 width = 300>

```python
print('-' * 30)
print('달러($)  원화(원)  유로(€)')
print('-' * 30)

for i in range(10, 101, 10) :
  dollar = i
  kr = i * 1080
  eur = i * 0.81
  print('  %d      %d    %.1f' %(dollar, kr, eur))

print('-' * 30)
```

### 10. while문을 이용하여 센터미터(cm), 밀리리터(mm), 미터(m), 인치(inch)의 길이 환산표를 만드는 프로그램을 작성하시오.

<p>* 1cm =: 10mm, 0.01m, 0.3937inch</p>

<p><실행 결과></p>
<img src = "https://drive.google.com/uc?id=1ImND5-xtXB74-wWeTC5kg3OICl6M5qBX" height = 400 width = 300>

```python
print('-' * 30)
print('cm  mm  m  inch')
print('-' * 30)

cm = 1
mm = 10
m = 0.01
inch = 0.3937

while cm <= 100:
  mm = cm * 10
  m = cm * 0.01
  inch = cm * 0.3937

  print('%d  %d  %.2f  %.2f' %(cm, mm, m, inch))

  cm += 2

print('-' * 30)
```
