# **4. 반복문**

### **4.1 반복문이란?**

### 예제 4-1. for문으로 반복 출력하기

````python
for x in range(5) :
    print('안녕하세요')
````

### 예제 4-2. while문으로 반복 출력하기

````python
x = 0
while x < 5 :
    print('안녕하세요')
    x += 1
````

### **4.2 for문**

### 4.2.1 for문의 기본 구조

### 예제 4-3. 1~10의 합계(for문)

````python
sum = 0

for i in range(1, 11) :
    sum += i
    print('i의 값 : %d, 합계 : %d' % (i, sum))
````

### 4.2.2 for문에서 range() 함수 활용



```
range(종료값)
 - 0 종료값 -1까지

range(시작값, 종료값)
 - 시작값부터 종료값 -1

range(시작값, 종료값, 증가값 or 감소값)
 - 시작값부터 종료값 -1까지 증가값만큼 변화
```



### 예제. 4-4 range() 함수의 활용

````python
for i in range(10) :
    print(i, end = ' ')
print()

for i in range(1, 11) :
    print(i, end = ' ')
print()

for i in range(1, 10, 2) :
    print(i, end = ' ')
print()

for i in range(20, 0, -2) :
    print(i, end = ' ')
````

### 4.2.3 배수의 합계

### 예제 4-5. 3이 배수의 합계

````python
sum = 0

for i in range(1, 101) :
    if i % 3 == 0 :
        print('%d' % i, end = '')
        sum += i

print('\n', '-' * 50)
print('1~100에서의 3의 배수의 합계 : %d' % sum)
````

```
자주 사용되는 이스케이프 코드
 - \n : 줄바꿈
 - \t : 탭
 - \\ : 역 슬래쉬(\) 자체를 출력
 - \' : 단 따옴표(')를 출력
 - /" : 쌍 따옴표(")를 출력
```

### 4.2.5 단위 환산표 만들기

### 예제 4-7 섭씨 온도를 화씨로 변환

````python
# 화씨 온도 = 섭씨 온도 * 9/5 + 32

print('-' * 30)
print('   섭씨    화씨')
print('-' * 30)

for c in range(-20, 31, 5) :
    f = c * 9.0/5.0 + 32.0
    print('%8d %6.1f' % (c, f))

print('-' * 30)
````

### **4.3 이중 for문**

### 예제 4-8. 2단 구구단표 만들기

````python
a = 2

for b in range(1, 10) :
    c = a * b
    print('%d x %d = %d' % (a, b, c))
````

### 예제 4-9. 전체 구구단표 만들기

````python
print('-' * 50)

for a in range(2, 10) :
    for b in range(1, 10) :
        c = a * b
        print('%d x %d = %d' % (a, b, c))
    print('-' * 50)
````

### **4.4 while문**

### 4.4.1 while문의 기본구조

### 예제 4-10. 1~10의 합계(while문)

````python
sum = 0
i = 1

while i <= 10 :
    sum += i
    print('i의 값 : %d => 합계 : %d' % (i, sum))
    i += 1
````



```
while문의 사용 형식

(1) 변수 값 초기화
(2) while 조건식 :
(3)     문장1
        문장2
        ....
(4) 변수 값의 증가(또는 감소)
```



### 4.4.2 배수의 합계(while문)

### 예제 4-11. 배수의 합계 구하기(while문)

````python
n1 = int(input('첫 수를 입력하세요 : '))
n2 = int(input('끝 수를 입력하세요 : '))
n = int(input('합계를 구하고자 하는 배수를 입력하세요 : '))

sum = 0
i = n1                # 첫 수 입력

while i < n2 + 1 :    # 100 < 300 + 1
    if i % n == 0 :   # 100 % 5 == 0 => 나머지가 0 즉 n의 배수
        sum += i      # sum = sum + i

    i += 1            # i = i + 1

print('%d~%d까지의 정수 중 %d의 배수의 합계 : %d' % (n1, n2, n, sum))
````

### 4.4.3 while문에서 문자열 다루기

### 예제 4-12. 영어 모음과 개수 구하기

````python
s = input('영어 문장을 입력하세요 : ')

i = 0
count = 0

print('모음 : ', end = '')

while i <= len(s) - 1 :
    if(s[i] == 'a' or s[i] == 'A' or s[i] == 'e' or s[i] == 'E'\
       or s[i] == 'i' or s[i] == 'I' or s[i] == 'o' or s[i] == 'O'\
       or s[i] == 'u' or s[i] == 'U') :
        count += 1
        print(s[i], end=' ')

    i += 1

print('\n모음의 개수 : %d' % count)
````
