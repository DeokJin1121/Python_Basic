### Q3-1 월을 입력받아 계절 표시하기

### 문제 : 다음은 if~ 구문을 이용하여 월을 입력받아 해당 꼐절을 표시하는 프로그램이다. 밑줄 친 부분을 채우시오.

<p><실행결과></p>
<p>월을 입력하세요 : 7</p>
<p>7월은 여름입니다.</p>

```python
month = int(input('월을 입력하세요 : '))

if month >= 3 and month <= 5 :
    print('%d월은 봄입니다.' % month)

if month >= 6 and month <= 8 :
    print('%d월은 여름입니다.' % month)

if month >= 9 and month <= 11 :
    print('%d월은 가을입니다.' % month)

if month == 12 or month == 1 or month == 2 :
    print('%d월은 겨울입니다.' % month)
```

### Q3-4 영어 알파벳 모음/자음 판별하기

### 문제 : 다음은 if~ else~ 구문을 이용하여 알파벳 하나를 입력받아 모음인지 자음인지를 판별하는 프로그램이다. 밑줄 친 부분을 채우시오.


<p><실행결과></p>
<p>영어 알파벳을 입력하세요 : w</p>
<p>w은(는) 자음이다.</p>

```python
char = input('영어 알파벳을 입력하세요 : ')

char2 = char.upper()

if char2 == 'A' or char2 == 'E' or char2 == 'I' or char2 == 'O' or char == 'U' :
    print('%s은(는) 모음이다.' % char)
else :
    print('%s은(는) 자음이다.' % char)
```

### Q3-6 할인율에 따른 지불 금액 계산하기

### 문제 : 다음은 if~ elif~ else~ 구문을 이용하여 물건 구매가를 입력받아 할인율에 따라 지불 금액을 계산하는 프로그램이다. 밑줄 친 부분을 채우시오.

### * 할인율 : 10000원 이상 ~ 50000원 미만 : 5%, 50000원 이상 ~ 300000원 미만 : 7.5%, 300000원 이상 : 10%

<p><실행결과></p>
<p>물 구매가를 입력하세요 : 60000</p>
<p>구매가 : 60000원</p>
<p>할인율 : 7.5%</p>
<p>할인 금액 : 4500원</p>
<p>지불 금액 : 55500원</p>

```python
buy = int(input('물건 구매가를 입력하세요 : '))

if buy >= 10000 and buy < 50000 :
    rate = 5
elif buy >= 50000 and buy < 300000 :
    rate = 7.5
elif buy >= 300000 :
    rate = 10

else :
    rate = 0

discount = buy * rate / 100
pay = buy - discount

print('구매가 : %.0f원' % buy)
print('할인율 : %.1f%%' % rate)
print('할인 금액 : %.0f원' % discount)
print('지불 금액 : %.0f원' % pay)
```

### Q3-7 세 수 중 가장 큰 수 찾기

### 문제 : 다음은 if~ elif~ else~ 구문을 이용하여 입력받은 세 수 중 가장 큰 수를 찾는 프로그램이다. 밑줄 친 부분을 채우시오.

<p><실행결과></p>
<p>첫 번째 정수를 입력하세요 : 34</p>
<p>두 번째 정수를 입력하세요 : 33</p>
<p>세 번째 정수를 입력하세요 : 27</p>
<p>입력된 세 수 34, 33, 27 중에서 가장 큰 수는 34 입니다.</p>

```python
num1 = int(input('첫 번째 정수를 입력하세요 : '))
num2 = int(input('두 번째 정수를 입력하세요 : '))
num3 = int(input('세 번째 정수를 입력하세요 : '))

if (num1 >= num2) and (num1 >= num3) :
    largest = num1
elif (num2 >= num1) and (num2 >= num3) :
    largest = num2
else :
    largest = num3

print('입력된 세 수 %d, %d, %d 중에서 가장 큰 수는 %d 입니다.' % (num1, num2 , num3, largest))
```

### Q3-8 고객 만족에 따른 팁 계산하기

### 문제 : 다음은 음식점 직원 서비스에 대한 고객 만족도에 따라 팁을 계산하는 프로그램이다. 밑줄 친 부분을 채우시오.

### * 팁 : 매운만족(1) 20%, 만족(2) 10%, 불만족(3) 5%

<p><실행결과></p>
<p>서비스 만족도 :</p>
<p>  1 : 매우만족</p>
<p>  2 : 만족7</p>
<p>  3 : 불만족</p>
<p>서비스 만족도를 입력해주세요(예 : 1 또는 2 또는 3) : 2</p>
<p>음식값을 입력해 주세요(예:8000) : 9000</p>
<p>서비스 만족도 : 만족, 팁 : 900원</p>

```python
print('서비스 만족도 : ')
print('  1 : 매우만족')
print('  2 : 만족')
print('  3 : 불만')
a = input('서비스 만족도를 입력해주세요(예 : 1또는 2 또는 3) : ')

price = int(input('음식값을 입력해 주세요(예 : 8000): '))

if a == '1' :
    tip = int(price * 0.2)
    service = '매우 만족'
elif a == '2':
    tip = int(price * 0.1)
    service = '만족'
else :
    tip = int(price * 0.05)
    service = '불만족'

print()
print('서비스 만족도 : %s, 팁 : %d원' % (service, tip))
```
