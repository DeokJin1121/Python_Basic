### 3.1.1 if문의 동작 원리

### 예제 3-1. if~ else~ 구문으로 양수 판별

````python
x = int(input('숫자를 입력하세요 : '))

if x > 0 :
    print('양수!')
else :
    print('0 또는 음수!')
````

### 예제 3-2. 비교 연산자의 참/거짓

### 예제 3-3. 짝수/홀수 판별

````python
num = int(input('숫자를 입력하세요 : '))

if num % 2 == 0 :
    print('짝수이다.')
else :
    print('홀수이다.')
````

### 예제 3-4. 자격증 시험 합격/불합격 판정 : and

````python
score1 = int(input('필기성적을 입력하세요 : '))
score2 = int(input('실기성적을 입력하세요 : '))

if score1 >= 80 and score2 >= 80 :
    print('합격!')
else :
    print('불합격!')
````

### 예제 3-5. 홈페이지 관리자 판별 : or

````python
id = input('아이디를 입력하세요 : ')
level = int(input('회원 레벨을 입력하세요 : '))

if id == 'admin' or level == 1 :
    print('관리자이다.')
else :
    print('관리자가 아니다.')
````

### 예제 3-6. 논리 연산자 not의 사용 예 : not

````python
name = input('이름을 입력하세요 : ')

if not name :
    print('이름이 입력되지 않았다.')
else :
    print('이름 : %s' % name)
````

### 예제 3-7. 참/거짓 판정

````python
print(True) if 15 else print(False)
print(True) if 0 else print(False)
print(True) if None else print(False)
print(True) if '홍길동' else print(False)
````

### 예제 3-9. if~ else~ 구문의 사용 예

````python
answer = '12345'
password = input('비밀번호를 입력하세요 : ')

if password == answer :
    print('비밀번호 OK!')
else :
    print('비밀번호 Not OK!')
````

### 예제 3-10. if~ elif~ else~ 구문_ 사용 예

````python
score = int(input('점수를 입력하세요 : '))

if score >= 90 :
    grade = 'A'
elif score >= 80 :
    grade = 'B'
elif score >= 70 :
    grade = 'C'
elif score >= 60 :
    grade = 'D'
else :
    grade = 'F'

print('성적 : %d점' % score)
print('등급 : %s' % grade)
````

### **3.6 if문으로 만 나이 계산하기**

### 예제 3-11. 만 나이 계산하기

````python
now_year = int(input('현재년을 입력해 주세요 : '))
now_month = int(input('현재월을 입력해 주세요 : '))
now_day = int(input('현재일을 입력해 주세요 : '))

birth_year = int(input('출생년을 입력해 주세요 : '))
birth_month = int(input('출생월을 입력해 주세요 : '))
birth_day = int(input('출생일을 입력해 주세요 : '))

if birth_month < now_month :                # 출생월이 현재월 이전일 경우
    age = now_year - birth_year
elif birth_month == now_month :             # 출생월이 현재월과 같은 경우
    if birth_day < now_day :                # 출생일이 현재일 이전일 경우
        age = now_year - birth_year
    else :                                  # 현재일이 출생일 이후일 경우
        age = now_year - birth_year - 1
else :                                      # 출생월이 현재월 이후일 경우
    age = now_year - birth_year - 1

print('-' * 50)
print('오늘 날짜 : %d년 %d월 %d일' % (now_year, now_month, now_day))
print('생년 월일 : %d년 %d월 %d일' % (birth_year, now_year, now_day))
print('-' * 50)
print('만 나이 : %d세' % age)
print('-' * 50)
````
