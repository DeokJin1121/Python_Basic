### Q7-1 코딩 연습 : 함수로 정수 합계 구하기

<p>문제 : 다음은 함수를 이용하여 정수의 합계를 구하는 프로그램이다. 밑줄 친 부분을 채우시오.</p>

<p><실행 결과></p>
<p>10 ~ 100의 정수 합계 : 5005</p>
<p>100 ~ 1000의 정수 합계 : 495550</p>
<p>1000 ~ 10000의 정수 합계 : 49505500</p>

```python
def sum(start, end) :
  total = 0
  for i in range(start, end + 1) :
    total += i
  print('%d ~ %d의 정수 합계 : %d' % (start, end, total))

sum(10, 100)
sum(100, 1000)
sum(1000, 10000)
```

### Q7-2 코딩 연습 : 함수로 배수 합계 구하기

<p>문제 : 다음은 함수를 이용하여 배수의 합계를 구하는 프로그램이다. 밑줄 친 부분을 채우시오.</p>

<p><실행 결과></p>
<p>시작 수를 입력하세요 : 10</p>
<p>끝 수를 입력하세요 : 100</p>
<p>합계를 구할 배수를 입력하세요 : 5</p>
<p>10 ~ 100의 정수 중 5의 배수의 합 : 1045</p>

```python
def sum_besu(n1, n2, num) :
  sum = 0
  for i in range(n1, n2 + 1) :
    if i % num == 0 :
      sum += i

  return sum

start = int(input('시작 수를 입력하세요 : '))
end = int(input('끝 수를 입력하세요 : '))
besu = int(input('합계를 구할 배수를 입력하세요 : '))

result = sum_besu(start, end, besu)

print('%d ~ %d의 정수 중 %d의 배수의 합 : %d' %(start, end, besu, result))
```

### Q7-3 코딩 연습 : 함수로 최대 공약수 구하기

<p>문제 : 다음은 함수를 이용하여 최대 공약수를 구하는 프로그램이다. 밑줄 친 부분을 채우시오.</p>

<p><실행 결과></p>
<p>첫 번째 수를 입력하세요 : 33</p>
<p>두 번재 수를 입력하세요 : 44</p>
<p>33과(와) 44의  최대공약수 : 11</p>

```python
def computeMaxGong(x, y) :
  if x > y :
    small = y
  else :
    small = x

  for i in range(1, small + 1):
    if ((x % i == 0) and (y % i == 0)) :
      result = i
  return result

num1 = int(input('첫 번째 수를 입력하세요 : '))
num2 = int(input('두 번재 수를 입력하세요 : '))

max_gong = computeMaxGong(num1, num2)

print('%d과(와) %d의  최대공약수 : %d' % (num1, num2, max_gong))
```

### Q7-4 코딩 연습 : 함수로 만드는 영어 단어 퀴즈

<p>문제 : 다음은 함수를 이용하여 영어 단어 맞추기 퀴즈를 만드는 프로그램이다. 밑줄 친 부분을 채우시오.</p>

<p><실행 결과></p>
<p>오렌지에맞는 영어 단어는? orange</p>
<p>맞습니다!</p>
<p>과자에맞는 영어 단어는? cookie</p>
<p>맞습니다!</p>
<p>어머니에맞는 영어 단어는? motheer</p>
<p>틀렸습니다!</p>
<p>형제에맞는 영어 단어는? brother</p>
<p>맞습니다!</p>
<p>파이썬에맞는 영어 단어는? pithon</p>
<p>틀렸습니다!</p>

```python
def matchWord(word, answer) :
  if word == answer :
    msg = '맞습니다!'
  else :
    msg = '틀렸습니다!'

  return msg

eng_dict = {'orange':'오렌지', 'cookie':'과자', 'mother':'어머니', 'brother':'형제', 'python':'파이썬'}

for key in eng_dict :
  string = input(eng_dict[key] + '에맞는 영어 단어는? ')
  result = matchWord(string, key)
  print(result)
```

### Q7-5 코딩 연습 : 함수로 세 수 중 큰 수 찾기

<p> 문제 : 다음은 함수를 이용하여 세 개의 정수 중에서 가장 큰 수를 찾는 프로그램이다. 밑줄 치 부분을 채우시오.</p>

<p><실행 결과></p>
<p>첫 번째 수를 입력하세요 : 33</p>
<p>두 번째 수를 입력하세요 : 77</p>
<p>세 번째 수를 입력하세요 : -21</p>
<p>33, 77, -21 중 가장 큰 수 : 77</p>

```python
def maxTwo(i, j) :
  if i > j :
    return i
  else :
    return j

def maxThree(x, y, z) :
  max1 = maxTwo(x, y)
  max2 = maxTwo(y, z)

  if max1 > max2 :
    largest = max1
  else :
    largest = max2
  return largest

a = int(input('첫 번째 수를 입력하세요 : '))
b = int(input('두 번째 수를 입력하세요 : '))
c = int(input('세 번째 수를 입력하세요 : '))

max_num = maxThree(a, b, c)
print('%d, %d, %d 중 가장 큰 수 : %d' % (a, b, c, max_num))
```

### Q7-6 코딩 연습 : 파일에서 성적 합계/평균 구하기

<p> 문제 : 다음은 예제 7-15에서 생성한 scores.txt 파일에서 성적 데이터를 읽어서 합계와 평균을 구하는 프로그램이다. 밑줄 친 부분을 채우시오.</p>

<p><실행 결과></p>
<p>--------------------------------------------------</p>
<p>김소영</p>
<p>합계 : 533, 평균 : 88.83</p>
<p>--------------------------------------------------</p>
<p>정예린</p>
<p>합계 : 532, 평균 : 88.67</p>
<p>--------------------------------------------------</p>
<p>이세영</p>
<p>합계 : 517, 평균 : 86.17</p>
<p>--------------------------------------------------</p>
<p>정수정</p>
<p>합계 : 543, 평균 : 90.50</p>
<p>--------------------------------------------------</p>
<p>박지수</p>
<p>합계 : 538, 평균 : 89.67</p>
<p>--------------------------------------------------</p>

```python
# scores.txt 파일 생성
scores = ['김소영 82 80 93 97 93 88',
        '정예린 86 100 93 86 90 77',
        '이세영 91 88 99 79 92 68',
        '정수정 86 100 93 89 92 83',
        '박지수 80 100 95 89 90 84']

data = ''

for item in scores :
  data += item + '\n'

# 화면 출력하기
print(data)

# 파일(scores.txt)에 저장하기
file = open('scores.txt', 'w', encoding = 'utf8')
file.write(data)
file.close()
```

```python
file = open('scores.txt', 'r', encoding = 'utf8')
lines = file.readlines()
file.close()

print('-' * 50)

for line in lines :
  student = line.split()
  i = 0
  sum = 0
  while i < len(student) :
    if i == 0 :
      print(student[i])
    else :
      sum += int(student[i])
    i += 1
  print('합계 : %d, 평균 : %.2f' % (sum, sum/6))
  print('-' * 50)
```
