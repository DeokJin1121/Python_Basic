### 7.1.2 함수의 종류

1.  사용자 함수

### 예제 7-2. 매개변수가 1개인 경우

````python
def even_odd(num):
  if num % 2 == 0:
    print('%d은(는) 짝수이다.' % num)
  else:
    print('%d은(는) 홀수이다.' % num)

even_odd(7)
even_odd(16)
````

### 예제 7-3. 매개변수가 여러개 인 경우

````python
def favorate_color(name, color, amount):
  if (amount == 1) :
    print('%s님은 %s을 좋아하지 않습니다.' % (name, color))
  elif (amount == 2) :
    print('%s님은 %s을 조금 좋아합니다.' % (name, color))
  else :
    print('%s님은 %s을 매우 좋아합니다.' % (name, color))

favorate_color('김지영', '빨강', 1)
favorate_color('홍채영', '노랑', 2)
favorate_color('진소진', '파랑', 3)
````

### 7.2.2 매개변수 *args

### 예제 7-4. 임의의 매개변수에서 사용되는 *

````python
def average(*scores) :
  sum = 0
  for i in range(len(scores)) :
    sum += scores[i]

    avg = sum / len(scores)
    print('%d과목의 평균 : %.2f' % (len(scores), avg))

average(80, 90, 100)
average(75, 80,94, 78)
average(80, 73, 76, 86, 82)
````

### 7.2.3 매개변수의 변수 값 전달 방식

1.   값에 의한 호출(Call by Value)
*   이 방식은 일반적인 함수 호출에서 일어나는 방식으로 호출 함수 측의 데이터나 변수를 정의한 함수 측의 매개변수에 복사한다. 숫자나 문자열 등이 전달될 대 흔히 일어나는 방식이다. 이 경우에는 정의된 함수에서 해당 매개변수 값을 변경하여도 호출한 메인 루틴에 있는 변수의 값은 변하지 않는다.

2.   레퍼런스에 의한 호출(Call by Reference)
*   파이썬의 리스트, 튜플, 딕셔너리(다른 언어에서의 배열)나 포인터(C나 C++ 언어에서 사용됨)를 매개변수로 전달 시에는 그 데이터가 저장된 메모리의 주소, 즉 레퍼런스가(Reference)가 매개변수에 전달된다. 데이터의 주소 값을 매개변수로 전달함으로써 정의된 함수에서 그 변수에 접근하여 값을 수정하거나 삭제할 수 있다. 이 때 호출된 함수에서 수정한 값은 함수를 호출한 메인 루틴에서도 그대로 적용된다.



### 예제 7-5. 값에 의한 호출의 예

```
id() 함수
 - 내장 함수인 id()는 변수를 포함하는 모든 객체가 저장된 컴퓨터 메모리의 주소를 나타내는 고유한 아이디를 얻는 데 사용된다.
```



````python
# 서브 루틴 : func() 함수
def func(x) :
  x = 100
  print('func() : x = ', x, ', id = ', id(x))

# 메인 루틴
x = 10
print('메인 : x = ', x, ', id = ', id(x))
func(x)
print('메인 : x =', x, 'id = ', id(x))
````

### 7.2.4 함수 값의 반환

### 예제 7-7. 함수 값의 반환의 예

```
# 함수의 반환 값이 사용되는 형식
① def 함수(매개변수1, 매개변수2, ...):
    문장1
    문장2
    ...
②  return 변수1
    ...
③  변수2 = 함수(입력값1, 입력값2, ...)
```



````python
def circle_area(r) :
  area = r * r * 3.14

  return area           # 함수 값의 반환

redius = int(input('원의 반지름을 입력하세요 : '))
result = circle_area(redius)
print('반지름 : %d, 원의 면적 : %.2f' % (redius, result))

redius = int(input('원의 반지름을 입력하세요 : '))
result = circle_area(redius)
print('반지름 : %d, 원의 면적 : %.2f' % (redius, result))
````

### 예제 7-9. 람다 함수의 다양한 사용 예

````python
f = lambda x, y, z : x + y + z
print(f(10, 20, 30))

def mul(n) :
  return lambda x : x * n

g = mul(3)                  # 10 * 3 = 30
h = mul(5)                  # 10 * 5 = 50

print(g(10))
print(h(10))
````

### 예제 7-13. 키워드 global의 사용 예

### TIP. 키워드 global
 - 키워드 global은 '지금부터 해당 변수를 전역 변수로 사용한다'는 것을 파이썬 인터프리터에게 알려주는 역할을 한다. 따라서 함수 정의 내부에서 키워드 global을 사용하면 함수 외부에서 사용 중인 변수 값을 현재의 범위, 즉 호출된 함수 내에서 변경할 수 있게 된다.



````python
def func() :
  global x
  x = 100
  print(x)
  print(id(x))

x = 10
print(x)
func()
print(x)
print(id(x))
````
