### Q4-1. 코딩 연습 : 5의 배수가 아닌 수의  합계 구하기

문제 : 다음은 키보드로 정부 범위의 시작 수와 끝 수를 입력받아 그 범위에 있는 수 중에서 5의 배수가 아닌 수의 합계를 구하는 프로그램이다. 밑줄 친 부분을 채우시오.

<p><실행 결과></p>
<p>시작 수를 입력하세요 : 100</p>
<p>끝 수를 입력하세요 : 200</p>
<p>---------------------------------------------------------</p>
<p>100에서 200까지의 5의 배수가 아닌 수의 합계 : 12000</p>

```python
n1 = int(input('시작 수를 입력하세요 : '))
n2 = int(input('끝 수를 입력하세요 : '))

sum = 0

for i in range(n1, n2 + 1) :
    if i % 5 != 0 :
        sum += i

print('-' * 50)
print('%d에서 %d까지 5의 배수가 아닌 수의 합계 : %d' % (n1, n2, sum))
```

### Q4-2 코딩 연습 : 전화번호에서 하이픈(-) 삭제하기

문제 : 다음은 하이픈(-)이 포함된 휴대폰 번호를 입력받아 하이픈을 삭제하는 프로그램이다. 밑줄 친 부분을 채우시오.

<p><실행 결과></p>
<p>하이픈(-)을 포함한 휴대폰 번호를 입력하세요 : 010-1234-5678</p>
<p>01012345678</p>

```python
number = input('하이픈(-)을 포함한 휴대폰 번호를 입력하세요 : ')

for x in number :                   # 문자열 하나씩 순회하면서 검사
    if x != '-' :
        print('%s' % x, end = '')
```

### Q4-3 코딩 연습 : 전화번호에 하이픈(-) 추가하기

문제 : 다음은 "###########" 형태의 11자리의 휴대폰 번호를 입력받아 하이픈(-)을 추가한 "###-####-####"의 형태로 전화번호를 출력하는 프로그램이다. 밑줄 친 부분을 채우시오.

<p><실행 결과></p>
<p>하이픈(-)을 뺀 11자리의 휴대폰 번호를 입력하세요 : 01033334444</p>
<p>010-3333-4444</p>

```python
phone = input('하이픈(-)을 뺀 11자리의 휴대폰 번호를 입력하세요 : ')

number = '';

for i in range(0, len(phone)) :
    if i == 2 :
        number = number + (phone[2] + '-')
    elif i == 6 :
        number = number + (phone[6] + '-')
    else :
        number = number + phone[i]

print(number)
```

### Q4-4 코딩 연습 : 길이 단위 환산표 만들기

문제 : 다음은 길이 단위(센티미티, 인치, 피트, 야드)에 대한 환산표를 만드는 프로그램이다. 밑줄 친 부분을 채우시오.

<p> * 환산 공식</p>
<p>inch = cm x 0.393701</p>
<p>ft = cm x 0.032808</p>
<p>yd = cm x 0.010936</p>

<p><실행 결과></p>
<p>-------------------------------------------------------</p>
<p>센티미터(cm) 인치(inch) 피트(ft) 야드(yd)</p>
<p>-------------------------------------------------------</p>
<p>      10        3.9       0.3       0.1</p>
<p>      20        7.9       0.7       0.2</p>
<p>      ...</p>
<p>      190       74.8      6.2       2.1</p>
<p>      200       78.7      6.6       2.2</p>
<p>-------------------------------------------------------</p>



```python
print('-' * 50)
print('센티미터(cm) 인치(inch)  피트(ft)  야드(yd)')
print('-' * 50)

for cm in range(10, 201, 10) :
    inch = cm * 0.393701
    ft = cm * 0.032808
    yd = cm * 0.010936
    print('%8d %10.1f %10.1f %10.1f' % (cm, inch, ft, yd))

print('-' * 50)
```

### Q4-5 코딩 연습 : 무게 단위 환산표 만들기

문제 : 다음은 길이 단위(킬로그램(kg), 파운드(lb), 온스(oz)에 대한 환산표를 만드는 프로그램이다. 밑줄 친 부분을 채우시오.

<p>* 환산 공식</p>
<p>lb = kg x 2.204623</p>
<p>oz = kg x 35.273962</p>

<p><실행 결과></p>
<p>--------------------------------------------------</p>
<p>킬로그램(kg) 파운드(lb)  온스(oz)</p>
<p>--------------------------------------------------</p>
<p>    10        22.0        352.7</p>
<p>    15        33.1        529.1</p>
<p>    20        44.1        705.5</p>
<p>    ...</p>
<p>    95        209.4       3351.0</p>
<p>    100       220.5       3527.4</p>
<p>--------------------------------------------------</p>

```python
print('-' * 40)
print('킬로그램(kg) 파운드(lb)  온스(oz)')
print('-' * 40)

for kg in range(10, 101, 5) :
    lb = kg * 2.204623
    oz = kg * 35.273962
    print('%8d  %10.1f  %10.1f' % (kg, lb, oz))

print('-' * 40)
```

### Q4-6 코딩 연습 : 홀수 개수 세기

문제 : 다음은 입력받은 숫자에 포함된 홀수의 개수를 세는 프로그램이다. 밑줄 친 부분을 채우시오.

<p><실행 결과></p>
<p>숫자를 입력하세요 : 3049894</p>
<p>입력된 숫자 중 홀수의 개수 : 3개</p>

```python
number = input('숫자를 입력하세요 : ')

total = 0

for a in number :
    a = int(a)
    if a % 2 == 1 :
        total += 1

print('입력된 숫자 중 홀수의 개수 : %d개' % total)
```

### Q4-7 코딩 연습 : 별표(*)로 특정 형태 만들기 1

문제 : 다음은 이중 for문을 이용하여 실행 결과가 같은 형태로 출력하는 프로그램이다. 밑줄 친 부분을 채우시오.

### Q4-9 코딩 연습 : 섭씨/화씨 환산표 만들기(while문)

문제 : 다음은 while문을 이용하여 섭씨 온도를 화씨 온도로 환산하는 표를 만드는 프로그램이다. 밑줄 친 부분을 채우시오.

<p><실행 결과></p>
<p>-------------------------------</p>
<p>    섭씨   화씨</p>
<p>-------------------------------</p>
<p>    -20    -4.0</p>
<p>    -19    -2.2</p>
<p>    -18    -0.4</p>
<p>    ...</p>
<p>     39    102.2</p>
<p>     40    104.0</p>
<p>-------------------------------</p>



```python
print('-' * 30)
print('    섭씨    화씨')
print('-' * 30)

c = -20
while c < 41 :
    f = c * 9.0/5.0 + 32.0
    print('%8d %8.1f' % (c, f))

    c += 1

print('-' * 30)
```

### Q4-10 코딩 연습 : 3의 배수가 안닌 수 출력하기(while문)

문제 : 다음은 while문을 이용하여 200~600의 수 중 3의 배수가 아닌 수를 출력하는 프로그램이다. 밑줄 친 부분을 채우시오.

<p>* 조건 한 줄에 8개씩 출력함</p>

<p><실행 결과></p>
<p>200 202 203 205 206 208 209 211</p>
<p>212 214 215 217 218 220 221 223</p>
<p>224 226 227 229 230 232 233 235</p>
<p>...</p>
<p>584 586 587 589 590 592 593 595</p>
<p>596 598 599</p>

```python
i = 200
count = 0
while i < 601 :
    if i % 3 != 0 :
        print(' %d' % i, end='')
        count += 1
        if count % 8 == 0 :
            print()
    i += 1
```
