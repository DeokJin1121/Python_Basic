### 1. for문을 이용하여 실행 결과와 같은 형태로 출력하는 프로그램을 작성하시오.

<p><실행 결과></p>
<p>* * * * * * * * * *</p>
<p>* * * * * * * * * *</p>
<p>* * * * * * * * * *</p>
<p>* * * * * * * * * *</p>
<p>* * * * * * * * * *</p>

```python
for i in range(1,50):
  if i % 10 == 0 :
    print()
  else :
    print('* ',end = '')
```

### 2. while문을 이용하여 1번 문제와 동일한 결과를 가져오는 프로그램을 작성하시오.

```python
i = 1

while i <= 50 :
  if i % 10 == 0 :
    print()
    i += 1
  else :
    print('* ', end ='')
    i += 1
```

### 3. for문을 이용하여 실행 결과와 같은 형태로 출력하는  프로그램을 작성하시오.

```python
for i in range(6, 1, -1) :
  print('%d ' % i * i)
```
