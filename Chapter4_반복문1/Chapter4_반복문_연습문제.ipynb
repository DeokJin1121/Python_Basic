### 1. for문을 이용하여 실행 결과와 같은 형태로 출력하는 프로그램을 작성하시오.

<p><실행 결과></p>
<p>* * * * * * * * * *</p>
<p>* * * * * * * * * *</p>
<p>* * * * * * * * * *</p>
<p>* * * * * * * * * *</p>
<p>* * * * * * * * * *</p>

```python
for i in range(1,50):
  if i % 10 == 0 :
    print()
  else :
    print('* ',end = '')
```

### 2. while문을 이용하여 1번 문제와 동일한 결과를 가져오는 프로그램을 작성하시오.

<p><실행결과></p>

<p>* * * * * * * * *</p>
<p>* * * * * * * * *</p>
<p>* * * * * * * * *</p>
<p>* * * * * * * * *</p>
<p>* * * * * * * * *</p>

```python
i = 1

while i <= 50 :
  if i % 10 == 0 :
    print()
    i += 1
  else :
    print('* ', end ='')
    i += 1
```

### 3. for문을 이용하여 실행 결과와 같은 형태로 출력하는  프로그램을 작성하시오.

<p><실행결과></p>

<p>5 5 5 5 5</p>
<p>4 4 4 4</p>
<p>3 3 3</p>
<p>2 2</p>
<p>1</p>

```python
for i in range(6, 0, -1) :
  print('%d ' % i * i)
```

### 4. for문을 이용하여 실행 결과와 같은 형태로 출력하는 프로그램을 작성하시오

<p><실행결과></p>

<p>0</p>
<p>0 1</p>
<p>0 1 2</p>
<p>0 1 2 3</p>
<p>0 1 2 3 4</p>
<p>0 1 2 3 4 5</p>

```python
x = ''

for i in range(0, 6) :
  x += (str(i) + ' ')
  print(x)
```

### 5. for문을 이용하여 실행 결과와 같은 형태로 출력하는 프로그램을 작성하시오.

<p><실행 결과></p>
<p>********</p>
<p>* &nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp *</p>
<p>* &nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp *</p>
<p>* &nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp *</p>
<p>* &nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp *</p>
<p>* &nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp *</p>
<p>* &nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp&nbsp *</p>
<p>********</p>

```python
for i in range(0, 8):
  if i == 0 or i == 7 :
    print('*' * 8)
  else :
    print('*      *')
```

### 6. 1 ~ 100까지의 수 중에서 홀수와 홀수의 합을 실행 결과와 같이 출력하는 프로그램을 작성하시오.

<p>1 + 3 + 5 + ... + 99</p>

<p><실행 결과></p>
<p>1 + 3 + 5 + 7 + 9 + ... + 95 + 97 + 99 = 2500</p>

```python
sum = 0

for i in range(1, 101) :
  if i % 2 == 1 and i != 99:
    print('%d + ' % i, end = '')
    sum += int(i)
  elif i == 99 :
    print('%d ' %i, end = '')
    sum += int(i)
  elif i == 100 :
    print('= %d' % sum)
```
