### 4. 온라인 서점에서 결제할 금액을 계산하는 프로그램을 작성하시오. 단, 실행결과는 아래와 정확하게 일치하도록 하고, 책 값, 할인율, 배송료는 키보드로 입력 받아 처리한다.

### * 힌트 : 결제 금약 = 책 값 - (책 값 * 할인율 / 100) + 배송료

<p><실행 결과></p>
<p>책 값을 입력하세요 : 15000</p>
<p>할인율 입력하세요(%) : 15</p>
<p>배송료를 입력하세요 : 3000</p>
<p>결제 금액 : 15750원</p>

````python
book_price = int(input('책 값을 입력하세요 : '))           # 책 값
discount_rate = int(input('할인율을 입력하세요(%) : '))      # 할인율
shipping_charges = int(input('배송료를 입력하세요 : '))   # 배송료

payment_amount = book_price - (book_price * discount_rate / 100) + shipping_charges

print('결제 금액 : %d 원' % payment_amount)
````

### 5. 이름, 현재년, 출생년을 입력받아 나이를 계산하는 프로그램을 작성하시오.
### * 힌트 : 나이 = 현재년 - 출생년 + 1

<p><실행 결과></p>
<p>이름을 입력하세요 : 안지영</p>
<p>현재년을 입력하세요 : 2020</p>
<p>탄생년을 입력하세요 : 1997</p>
<p>안지영님의 나이는 24세 입니다!</p>

````python
name = input('이름을 입력하세요 : ')
this_year = int(input('현재년을 입력하세요 : '))      # 현재년
year_of_birth = int(input('탄생년을 입력하세요 : '))  # 출생년

age = this_year - year_of_birth + 1

print('%s님의 나이는 %d세 입니다!' % (name, age))
````

### 6. 연, 월, 일을 입력받아 ####-##-## 형태로 출력하는 프로그램을 작성하시오. 단, 실행 결과에서와 같이 월과 일에 1 ~ 9의 숫자가 입력되면 그 숫자 앞에 0을 채운다.

<p><실행 결과></p>
<p>연을 입력하세요 : 2020</p>
<p>월을 입력하세요 : 1</p>
<p>일을 입력하세요 : 5</p>
<p>2020-01-05</p>

````python
year = int(input('연을 입력하세요 : '))
month = int(input('월을 입력하세요 : '))
day = int(input('일을 입력하세요 : '))

print('{:0>4d}-{:0>2d}-{:0>2d}'.format(year, month, day))
````
